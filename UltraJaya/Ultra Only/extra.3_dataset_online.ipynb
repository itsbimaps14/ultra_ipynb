```python
df_online5 = pd.read_csv('dataset_online/usable5_dataset_sa_ml/minuman.csv')
df_online5.reset_index(drop=True, inplace=True)
df_online5
```

```python
list_minuman = df_online5['id'].tolist()
print(list_minuman)
```

```python
minumal_all = pd.DataFrame(columns=['review', 'rating'])

for minuman in list_minuman:
  path = 'dataset_online/usable5_dataset_sa_ml/minuman/'+str(minuman)+'.csv'
  if os.path.exists(path):
    minuman_ = pd.read_csv(path)
    if len(minuman_) != 0:
      # print(str(minuman) +' - '+str(len(minuman_)))
      # print(minuman_)
      minumal_all = pd.concat([minumal_all, minuman_])
```

```python
minumal_all
```

```python
minumal_all = minumal_all[minumal_all['rating'] < 4]
minumal_all.reset_index(drop=True, inplace=True)
minumal_all.to_excel('dataset_online/filter/usable5_dataset_sa_ml.xlsx')
minumal_all
```
