## **STRUCTURE CHANGES TO MATCH TRAINING DATA REQUIREMENT**

```python
## Declare Hashtag Data
hashtag = "ultramilk_initialize"
```

```python
# Load Data for Training
df = pd.read_excel('data/preprocess/'+hashtag+'.xlsx')

# Rename column previous index as id_tweet for reporting purposes
df.rename( columns={'Unnamed: 0':'id_tweet'}, inplace=True)

# Show Data
df
```

```python
## DATA MENTIONLYTICS
# Load Data for Training
df_mention = pd.read_excel('data/preprocess/'+hashtag+'_mentionlytics.xlsx')

# Delete Unused Column
df_mention.drop(columns=['sentiment', 'predicted'], inplace=True)

# Add Missing Column
df_mention['date'] = '2000-01-01'
df_mention['platform'] = 'Mentionlytics'
df_mention['brand'] = 'Ultramilk'
df_mention['hashtag'] = '#Mentionlytics'
df_mention['name'] = 'Mentionlytics'
df_mention['username'] = '@mentionlytics'
df_mention['comments'] = 0
df_mention['share_retweet'] = 0
df_mention['bookmark'] = 0
df_mention['quotes'] = 0
df_mention['likes'] = 0
df_mention['link'] = '-'

# Rename Column
df_mention.rename(columns={'ID': 'profile_id', 'true_sen': 'labelling'}, inplace=True)

# Rearrange Column Loc
df_mention = df_mention[['date', 'platform', 'brand', 'hashtag', 'name', 'username', 'profile_id', 'comments', 'share_retweet', 'bookmark',
                         'quotes', 'likes', 'link', 'text', 'labelling', 'text_final']]

# Show Data
df_mention
```

## **CREATE FOLDER AND EXPORT DATA**

```python
# Check Path is Available if no Create Dir
if path.exists(root_file+'/data/training/'+hashtag) == False:
  os.makedirs(root_file+'/data/training/'+hashtag)
else :
  print("File Already Existed")
```

```python
# Export Data
df.to_csv('data/training/'+hashtag+'/'+hashtag+'.csv', index=False)

df_mention.to_csv('data/training/'+hashtag+'/'+hashtag+'_mention.csv', index=False)
```
