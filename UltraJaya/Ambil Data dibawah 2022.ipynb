```python
import re

text = "Visit my website at http://example.com or www.example.net. Check out instagram.com and acer.net. bima.suka ...ggg"

# Define the regex pattern
url_pattern = re.compile(r'\b(?:https?://|www\.)\S+\b|\b(?:\S+\.)?(?!http://|www\.)\S+\.\S+\b', re.IGNORECASE)

# Remove URLs that don't start with "http://" or "www."
filtered_text = re.sub(url_pattern, '', text)

print(filtered_text)
```

```python
## Install Library
!pip install lingua-language-detector
from lingua import Language, LanguageDetectorBuilder
languages = [Language.INDONESIAN, Language.ENGLISH]
detector = LanguageDetectorBuilder.from_languages(*languages).with_minimum_relative_distance(0.2).build()

import re

def get_lang_df(text):
  text = str(text)
  text = re.sub('@[^\s]+','',text)
  text = re.sub(r'http\S+', '', text)
  text = re.sub('#[^\s]+','',text)

  # Define the regex pattern
  url_pattern = re.compile(r'\b(?:https?://|www\.)\S+\b|\b(?:\S+\.)?(?!http://|www\.)\S+\.\S+\b', re.IGNORECASE)

  # Remove URLs that don't start with "http://" or "www."
  text = re.sub(url_pattern, '', text)

  try:
      language = detector.detect_language_of(text)
  except:
      language = "error"
  return language

def check_confidence_val(text):
  text = str(text)
  text = re.sub('@[^\s]+','',text)
  text = re.sub(r'http\S+', '', text)
  text = re.sub('#[^\s]+','',text)

  # Define the regex pattern
  url_pattern = re.compile(r'\b(?:https?://|www\.)\S+\b|\b(?:\S+\.)?(?!http://|www\.)\S+\.\S+\b', re.IGNORECASE)

  # Remove URLs that don't start with "http://" or "www."
  text = re.sub(url_pattern, '', text)

  confidence = detector.compute_language_confidence(text, Language.INDONESIAN)
  return f"{confidence:.4f}"
```

```python
# result = [get_language(x) for x in df_data_komen['text']]
df_merge['lang'] = df_merge['text'].apply(lambda x: get_lang_df(x))
df_merge['conf_score'] = df_merge['text'].apply(lambda x: check_confidence_val(x))
df_merge['conf_score'] = df_merge['conf_score'].astype(float)
```

```python
df_merge.reset_index(drop=True, inplace=True)
df_merge.to_excel('data_scraped/add_data/all_combined.xlsx')
df_merge
```
